```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_func(func, x, title = ''):
    y = func(x)
    x_len = np.max(x) - np.min(x)
    y_len = np.max(y) - np.min(y)
    
    plt.title(title)
    plt.axis([np.min(x) - 0.1*x_len, np.max(x) + 0.1*x_len, \
              np.min(y) - 0.1*y_len, np.max(y) + 0.1*y_len])
    plt.xlabel('x axis')
    plt.ylabel('y axis')
    plt.plot(x, y)
    plt.show()
```

```python
plot_func(lambda x: x**2, np.arange(0, 1, 0.01))
```

```python
def entropy(x): return -x*np.log(x)
```

```python
plot_func(entropy, np.arange(0.01, 1, 0.01))
```
